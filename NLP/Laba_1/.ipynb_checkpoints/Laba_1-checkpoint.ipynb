```python
import deeppavlov
print(deeppavlov.__version__)
```

```python
from deeppavlov import build_model, configs

model = build_model(configs.classifiers.rusentiment_convers_distilrubert_2L, download=False)

texts = [
    "Это отличная новость!",
    "Ужасная погода сегодня",
    "Нейтральное сообщение",
    "Я очень зол на это",
    "Какой прекрасный день!"
]

results = model(texts)

for text, sentiment in zip(texts, results):
    print(f"Текст: {text}")
    print(f"Тональность: {sentiment}")
    print("-" * 30)
```
