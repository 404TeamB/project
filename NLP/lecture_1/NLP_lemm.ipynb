Пример использования стемминга:

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
words = ["program", "programs", "programmer", "programming", "programmers"]
stemmed_text = [stemmer.stem(word) for word in words]
print(stemmed_text)
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("russian")
my_words = ['Василий', 'Геннадий', 'Виталий']
l=[stemmer.stem(word) for word in my_words]
print(l)
```

Пример создания "мешка слов"

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer()
```

```python
corpus = [
    'ехать Гpека чеpез pека',
    'видеть Гpека в pека pак',
    'сунуть Гpека pука в pека',
    'pак за pука Гpека цап'
]
```

```python
bow = count_vect.fit_transform(corpus)
```

```python
bow.shape
```

```python
print(bow.toarray())
```

Очистим мешок слов от стоп-слов:

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('russian'))
```

```python
count_vect = CountVectorizer(stop_words=list(stop_words))
```

```python
bow2 = count_vect.fit_transform(corpus)
```

```python
bow2.shape
```

```python
print(bow2.toarray())
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
count_tf_idf = TfidfVectorizer(stop_words=list(stop_words))
```

```python
tf_idf = count_tf_idf.fit_transform(corpus)
```

```python
print(tf_idf)
```

Передав TfidfVectorizer() аргумент ngram_range, можно рассчитать N-граммы.

```python
count_ngrams = TfidfVectorizer(ngram_range=(3,3))
```

```python
ngrams = count_ngrams.fit_transform(corpus)
```

```python
print(ngrams)
```
